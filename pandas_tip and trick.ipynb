# aggregate 



```python
import pandas as pd 
import seaborn as sns
df = sns.load_dataset('tips')
df.head()
```

```python
# Total number of passengers
print(df.shape[0])

# Average age of passengers
print(df['age'].mean())

# Max fare
print(df['fare'].max())

# Count non-null age values
print(df['age'].count())
```

```python
df.groupby('sex')['age'].mean()
```

```python
df.groupby('sex').agg({
    'age': ['mean', 'max', 'min'],
    'fare': ['mean', 'sum']
})
```

```python
df.agg({
    'age': ['mean', 'min', 'max'],
    'fare': ['sum', 'mean'],
    'survived': ['sum', 'mean']  # sum = number of survivors
})
```
